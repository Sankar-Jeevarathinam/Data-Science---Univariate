```python
import pandas as pd
dataset=pd.read_csv("kidney_disease.csv")
```

```python
from univariate import Univariate
```

```python
Quan,Qual=Univariate.QuanQual(dataset)
```

```python
Quan
```

```python
print(Qual)
```

```python
for columns in Quan:
    dataset[columns].fillna(dataset[columns].mean(), inplace=True)
```

```python
for columns in Qual:
    mode_value=dataset[columns].mode()[0]
    dataset[columns]=dataset[columns].fillna(mode_value)
```

```python
dataset.isnull().sum()
```

```python
descriptive=Univariate.univariate(dataset,Quan)
```

```python
descriptive
```

```python
lesser,greater=Univariate.outlier_check(Quan,descriptive)
```

```python
print("Lesser outlier in following columns:", lesser)
print("Greater outlier in following columns:",greater)
```

```python
Univariate.replace_lesser_outlier(lesser,dataset,descriptive)
```

```python
Univariate.replace_greater_outlier(lesser,dataset,descriptive)
```
